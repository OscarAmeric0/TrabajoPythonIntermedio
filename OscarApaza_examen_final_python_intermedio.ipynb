2. **(5 puntos)** Desarrolle una función que trabaje con argparse. Dicha función debe recibir como atributos debe recibir un tipo de animal y tratamiento y devolver como resultado el costo del servicio.

**Lista de precios**
> Corte de pelo
*   Gato: 20
*   Perro: 30
*   Conejo: 15

> Baño
*   Gato: 15
*   Perro: 20
*   Conejo: 10

> Desparasitación
*   Gato: 35
*   Perro: 50
*   Conejo: 25







```python
import argparse
import sys


def main():

    parser = argparse.ArgumentParser()
    parser.add_argument('--TipoMascota', type=str, default='null',
                        help='Es tipo de Mascota , el cual puede ser \
                        Gato,Perro,Conejo',
                        required=True)

    parser.add_argument('--Tratamiento', type=str,
                        help='Es el tipo de Tramiento que puede recibir \
                        la Mascota este es una lista separado por "," \
                        por ejemplo \'Corte_de_Pelo\', \'Baño\'',
                        required=True)

    args = parser.parse_args()
    sys.stdout.write(str(Costo_Servicio(args)))


def Costo_Servicio(args):
    Mascota = {'Gato': {'Corte_de_Pelo': 20, 'Baño': 15,
                        'Desparacitacion': 35},
               'Perro': {'Corte_de_Pelo': 30, 'Baño': 20,
                         'Desparacitacion': 50},
               'Conejo': {'Corte_de_Pelo': 15, 'Baño': 10,
                          'Desparacitacion': 25}}

    TipoMascotaPermitido = "\n".join(Mascota)

    if args.TipoMascota not in Mascota:
        return (f'El tipo Mascota {args.TipoMascota} no esta permitido,'
                f'solo se permite :\n{TipoMascotaPermitido}')

    Tratamientos = set(args.Tratamiento.split(","))
    CostServ = 0
    Servicios = ""

    try:
        for t in Tratamientos:
            CostServ += Mascota[args.TipoMascota][t]
            Servicios += (f'{t} por un valor de s/.'
                          f'{Mascota[args.TipoMascota][t]:,.2f}\n')

        print(f'Tratamiento para Mascota Tipo {args.TipoMascota}:\n'
              f'{Servicios}Total Costo del Servicio s/.{CostServ:,.2f}\n')
    except KeyError:
        print("El Tipo de Tratamiento Ingresado no es valido")


if __name__ == '__main__':
    main()
```

3. **(10 puntos)** Descargue la base de datos de casos positivos de covid del siguiente [enlace](https://www.datosabiertos.gob.pe/dataset/casos-positivos-por-covid-19-ministerio-de-salud-minsa) y carguelos en pandas, luego de esto grafique un gráfico de lineas que muestre la cantidad de casos mes a mes de positivos de covid en la región de Lima.




Captura de Pantalla 2023-03-04 a la(s) 21.24.04.png


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
covid19 = pd.read_csv('positivos_covid.csv',
                      delimiter=';')
```

```python
def contagios_por_departamento(dataset, departamento, titulo_grafico):
    covid19 = dataset.dropna()
    covid19_departamento = covid19.loc[covid19['DEPARTAMENTO'] == departamento]
    covid19_departamento['PERIODO'] = (
        covid19_departamento['FECHA_RESULTADO']
        .apply(lambda x: str(x)[0:4]+'-'+str(x)[4:6]))
    df = (covid19_departamento.groupby("PERIODO")[["id_persona"]].
          aggregate(['count']).reset_index())
    df.columns = df.columns.droplevel(-1)
    plt.rcParams["figure.figsize"] = (20, 6)
    plt.title(titulo_grafico+' '+departamento, fontsize=20)
    sns.lineplot(x="PERIODO", y="id_persona", data=df)
    plt.xticks(rotation=90)


contagios_por_departamento(covid19, 'LIMA', 'NUMERO DE CONTAGIADOS POR MES')
```

4. **(3 puntos)** En el dataset de Titanic train [enlace](https://www.kaggle.com/c/titanic/overview) visto en clase, realice un gráfico de cajas sobre la variable Fare (Costo del ticket), ¿qué observa?, ¿Hay outliers?, ¿Qué tratamiento recomendaría?

```python
titanic = pd.read_csv('train_titanic.csv')
titanic_group = (titanic.groupby('Pclass').
                 agg({'Fare': ['min', 'max', 'mean',
                               'median', 'count',
                               lambda x: x.quantile(0.25),
                               lambda x: x.quantile(0.5),
                               lambda x: x.quantile(0.75)]}))

titanic_group.columns = (['_'.join(col).strip() for
                          col in titanic_group.columns.values])
titanic_group
```

```python
FareMin = titanic.Fare.min()
FareMax = titanic.Fare.max()
plt.figure(figsize=(20, 8))
plt.xticks(np.arange(FareMin, FareMax, 15))
sns.set_theme(style="whitegrid")

sns.boxplot(data=titanic, x="Fare", y="Pclass", orient="h")
```

1. ¿qué observa? Se Observa que los pasajes en las 3 clases varian.
 *   En la 1ra Clase el valor maximo es 512.3292, el minimo 0 y su mediana es 60.2875, de un total de 216 pasajeros.
 *   En la 2da Clase el valor maximo es 73.5, el minimo 0 y su mediana es 14.25, de un total de 184 pasajeros.
 *   En la 3ra Clase el valor maximo es 69.55, el minimo 0 y su mediana es 60.2875, de un total de 491 pasajeros.

2. ¿Hay outliers? Se observa que en las 3 Clases de Pasajeros existen outliers , valores que esta fuera de los bigotes de cada bloxplot.

3. ¿Qué tratamiento recomendaría?
   Un tratamiento que recomendaría seria tomar los datos hasta los márgenes en los que se encuentra los bigotes, o hacer uso del atributo "showfliers" para no mostrar los outliers , el cual se establecería en "False".

```python
titanic1 = titanic.loc[(titanic.Fare >= 0) & (titanic.Fare <= 1000)]
plt.figure(figsize=(20, 8))
plt.xticks(np.arange(0, 520, 5))
sns.set_theme(style="whitegrid")

sns.boxplot(data=titanic1, x="Fare", y="Pclass", orient="h", showfliers=False)
```
